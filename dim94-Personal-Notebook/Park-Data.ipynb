### Introduction to the data

This data set lists all parks and trails located in the city of Pittsburgh. The data contains important information such as the **park names**, **total area**, and **physical locations**. The set contains much more information, but these three categories are the most important in order to determine the best neighborhood for dogs. 

```python
import pandas as pd

#prints info of first 5 parks in the set
parks = pd.read_csv("park-data.csv", sep=",")
parks.head(5)
```

---
### The Metric

Unlike air quality or crime data, **parks cannot be measured as a value or number**, but rather just as a physical location and information regarding each park as can be seen above. Therefore, there is no ordering of the data set that clearly chooses a best neighborhood. Of course, one way to put some kind of rank on all of the parks would be to sort them from **greatest to lowest area** in acres:

```python
#sort all parks by total areage
parks_area_sorted = parks.sort_values(by='acreage', ascending=False)

#select only the name and acreage and print the top 10 parks  
print(parks_area_sorted[['origpkname', 'acreage']].head(10))
```

Here is the same 10 parks and their areas in a bar graph to help visualize their sizes:

```python
import matplotlib.pyplot as plt

# Extract the data
top_10_parks = parks_area_sorted[['origpkname', 'acreage']].head(10)


# Create the bar graph
plt.figure(figsize=(9, 5))  # Adjust the size of the plot
plt.bar(top_10_parks['origpkname'], top_10_parks['acreage'], color='green')

# Add labels and title
plt.xlabel('Park Name')
plt.ylabel('Acreage')
plt.title('Top 5 Parks by Acreage')
plt.xticks(rotation=50, ha='right')  # Rotate the park names for better visibility

plt.show()
```

---
### Best Neighborhood
This ordering shows that **Hays Woods** is the largest park in area of the ones in this set. Any neighborhood bordering Hays Woods would be ideal for dogs, and **New Homestead**, which does in fact border Hays Woods, is a good choice according to the other groups of data. Another neighborhood that could be considered as best if solely using park data would be **Squirrel Hill** since it falls directly in between Frick and Schenley parks, ranked 2nd and 3rd largest respectively. However, Squirrel Hill did not score as highly in the other two data sets leaving New Homestead as the better choice.
